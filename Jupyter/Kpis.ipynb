# <center> **Kpis de Siniestros Viales**

Los KPIs, o Indicadores Clave de Proceso, son herramientas cruciales que evalúan la eficacia en la consecución de objetivos fundamentales. Estos indicadores representan valores que cuantifican el logro de metas, generalmente expresadas en índices o porcentajes. A diferencia de las métricas, que son mediciones de datos en sí mismas sin un objetivo específico, los KPIs se distinguen al medir metas clave específicas, no cualquier aspecto. Además, los KPIs requieren métricas para su cálculo, pero no todas las métricas son indicadores, y no todos los indicadores son KPIs.

# ***Primer KPI***

**Reducir en un 10% la tasa de homicidios en siniestros viales de los últimos seis meses, en CABA, en comparación con la tasa de homicidios en siniestros viales del semestre anterior**

Definimos a la tasa de homicidios en siniestros viales como el número de víctimas fatales en accidentes de tránsito por cada 100,000 habitantes en un área geográfica durante un período de tiempo específico. Su fórmula es: (Número de homicidios en siniestros viales / Población total) * 100,000

Se realiza Web Scraping para encontrar la poblacion de CABA,en los años requeridos.


Para el año 2016 se cuenta con una Poblacion de 3.059.122 (chrome-extension://efaidnbmnnnibpcajpcglclefindmkaj/https://www.argentina.gob.ar/sites/default/files/siempro_-_caba_10.07.17.pdf)

No se encontro informacion de los años 2017 y 2018, se procede a hacer un promedio de los años 2016 y 2019 para trabajar con estos datos, dando como resultado 3.065.575 habitantes.

Para el año 2019 se cuenta con una Poblacion de 3.072.029 (chrome-extension://efaidnbmnnnibpcajpcglclefindmkaj/https://www.argentina.gob.ar/sites/default/files/191202_caba.pdf)

Para el año 2020 habian 3.075.646 habitantes (https://www.argentina.gob.ar/sites/default/files/2021/09/reporte_caba_final.pdf)  

Se hace un promedio entre las dos poblaciones para encontrar la poblacion del 2021 y se obtiene una poblacion estimada de 3.098.676

Para el año 2022 un aproximado de 3.121.707 (https://www.argentina.gob.ar/pais/poblacion/centros). 
 

Creamos un diccionario con la poblacion por año

```python
poblacion_por_año = { 2016: 3059122,
    2017: 3065575 ,
    2018: 3065575 ,
    2019: 3072029,
    2020: 3075646,
    2021: 3121707}
```

```python
#Creamos la tabla con los datos requeridos en el kpi

kpi1= data.groupby(["Año", "Semestre"])["# Victimas"].sum().reset_index()

#Introducimos el diccionario de poblacion que creamos anteriormente
kpi1["Poblacion"] = kpi1["Año"].map(poblacion_por_año)

# Calculamos la tasa de homicidios
kpi1["Tasa"] = ((kpi1["# Victimas"] / kpi1["Poblacion"]) * 100000).round(2)

# Revisamos en términos porcentuales los cambios
kpi1["Cambio Porcentual"] = (kpi1["Tasa"].pct_change() * 100).round(2)


kpi1
```

Revisamos en que año y semestre si se cumplio con el kpi

```python
kpi1[kpi1["Cambio Porcentual"]<-10].reset_index()
```

Aunque se observaron disminuciones en la tasa de homicidios en siniestros viales en algunos periodos, solo en ciertos casos se logró cumplir con el KPI de reducir en un 10% la tasa en comparación con el semestre anterior.

Año 2017:

Tanto en el primer como en el segundo semestre, se experimentaron reducciones, pero ninguna alcanzó el 10%.

Año 2018:

Similar al año 2017, hubo reducciones en ambos semestres, pero ninguna cumplió con el KPI del 10%.

Año 2019:

Ambos periodos mostraron disminuciones de más del 20%, superando significativamente el objetivo del KPI.

Año 2020:

En el primer semestre de 2020, se cumplió el KPI, pero en el segundo semestre, la tasa aumentó en un 80.65%, incumpliendo el objetivo.

Año 2021:

Aunque en el primer semestre se logró una disminución del 3.23%, en el segundo semestre la tasa disminuyó en un 23.64%, cumpliendo con el KPI.
En resumen, si bien algunos periodos mostraron mejoras en la seguridad vial, solo en los años 2019 y 2021 se lograron reducciones consistentes de más del 10%, cumpliendo así con el objetivo establecido por el KPI.

```python
datakpi1 = kpi1.groupby(["Año", "Semestre"])[["# Victimas", "Poblacion", "Cambio Porcentual"]].sum().reset_index()
datakpi1

datakpi1.to_csv('datakpi1.csv',index=False, encoding='utf-8')
print(f'Se guardó el archivo ')
```

#  ***Segundo KPI***

**Reducir en un 7% la cantidad de accidentes mortales de motociclistas en el último año, en CABA, respecto al año anterior.**

Definimos a la cantidad de accidentes mortales de motociclistas en siniestros viales como el número absoluto de accidentes fatales en los que estuvieron involucradas víctimas que viajaban en moto en un determinado periodo temporal. Su fórmula para medir la evolución de los accidentes mortales con víctimas en moto es: (Número de accidentes mortales con víctimas en moto en el año anterior - Número de accidentes mortales con víctimas en moto en el año actual) / (Número de accidentes mortales con víctimas en moto en el año anterior) * 100

```python
#Buscamos los datos requeridos para el Kpi que es accidentes mortales en motocolistas

requerimientos= data["Victima"] == "MOTO"
kpi2 =data[requerimientos]


kpi2=  kpi2.groupby("Año")["# Victimas"].sum().reset_index()

#Encontramos la evolucion porcentual de las victimas 
kpi2['Evolucion Porcentual kpi2'] = (((kpi2['# Victimas'].shift(1) - kpi2['# Victimas']) /kpi2['# Victimas'].shift(1)) * 100).fillna(0).round(2)

kpi2
```

El KPI propuesto tiene como objetivo reducir en un 7% la cantidad de accidentes mortales de motociclistas en CABA en el último año en comparación con el año anterior. A continuación, se presenta un análisis de los resultados obtenidos:

**Cumplimiento del KPI:**

En los años 2016 y 2017, la evolución porcentual fue del 7.46%, superando el objetivo del 7% de reducción, lo que indica un cumplimiento positivo del KPI en esos años.

**Años 2018 y 2019:**

En 2018, la evolución porcentual fue del 1.61%, lo que sugiere una reducción, pero menor al 7% requerido por el KPI. Sin embargo, en 2019, se observa una evolución porcentual del 18.03%, superando significativamente el objetivo.

**Año 2020:**

En 2020, se destaca una disminución drástica del 40.00%, cumpliendo ampliamente con el KPI y demostrando una reducción sustancial en la cantidad de accidentes mortales de motociclistas.

**Año 2021:**

A pesar de la reducción lograda en 2020, en 2021 se observa una evolución porcentual negativa del -53.33%, indicando un aumento significativo en la cantidad de accidentes mortales. Esta variación negativa no cumple con el KPI y sugiere la necesidad de una revisión de las estrategias implementadas.



Es esencial analizar las tendencias a lo largo de los años para comprender mejor los factores que contribuyen a las variaciones en la evolución porcentual.
La reducción excepcional en 2020 podría estar influenciada por la pandemia de COVID-19 y las restricciones asociadas, lo que llevó a una disminución general en la movilidad.
El aumento abrupto en 2021 destaca la importancia de monitorear continuamente las condiciones del tráfico y de seguridad vial, ajustando las estrategias según sea necesario.
En resumen, mientras que se logró una reducción significativa en 2020, la evolución negativa en 2021 subraya la importancia de mantener un enfoque proactivo en las medidas de seguridad vial para abordar los desafíos cambiantes en la movilidad.

```python
datakpi2 = kpi2.groupby(["Año"])[[ "Evolucion Porcentual kpi2"]].sum().reset_index()
datakpi2

datakpi2.to_csv('datakpi2.csv',index=False, encoding='utf-8')
print(f'Se guardó el archivo ')
```

#  ***Tercer KPI***

**Reducir en un 5% la cantidad de accidentes mortales ocurridos en las avenidas en el último año, en CABA, respecto al año anterior.**

Definimos a la cantidad de accidentes mortales ocurridos en avenidas como el número absoluto de accidentes fatales en las avenidas en un determinado periodo temporal. Su fórmula para medir la evolución de los accidentes mortales en avenidas  es: (Número de accidentes mortales en avenidas en el año anterior - Número de accidentes mortales en avenidas en el año actual) / (Número de accidentes mortales en avenidas en el año anterior) * 100

```python
#Buscamos los datos requeridos para el Kpi que es accidentes mortales en motocolistas

requerimientos= data["Tipo de calle"] == "AVENIDA"
kpi3 =data[requerimientos]


kpi3=  kpi3.groupby("Año")["# Victimas"].sum().reset_index()

#Encontramos la evolucion porcentual de las victimas 
kpi3['Evolucion Porcentual kpi3'] = (((kpi3['# Victimas'].shift(1) - kpi3['# Victimas']) /kpi3['# Victimas'].shift(1)) * 100).fillna(0).round(2)

kpi3
```

La meta de reducir en un 5% la cantidad de accidentes mortales en avenidas en el último año, en la Ciudad Autónoma de Buenos Aires (CABA), en comparación con el año anterior. Los resultados muestran que en los años 2016 y 2017, no hubo cambios en la cantidad de víctimas en avenidas. En 2018, se registró una disminución del -8.60%, superando la meta del KPI. Sin embargo, en 2019, se observó un aumento del 34.65%, contraviniendo la meta establecida. El año 2020 destacó por una disminución del 15.15%, cumpliendo con la meta del KPI. En 2021, se evidenció una disminución del -8.93%, nuevamente superando la meta del KPI. En conjunto, aunque se alcanzaron reducciones notables en 2020 y 2021, es crucial analizar los factores que contribuyeron a las fluctuaciones, especialmente el aumento en 2019, para ajustar estrategias y mantener una mejora sostenida en la seguridad vial en las avenidas de CABA.

```python
datakpi3 = kpi3.groupby(["Año"])[[ "Evolucion Porcentual kpi3"]].sum().reset_index()
datakpi3

datakpi3.to_csv('datakpi3.csv',index=False, encoding='utf-8')
print(f'Se guardó el archivo ')
```
